# 01. Load the Dataset

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
data = pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

# 02. Data Cleaning

```python
data.isnull().sum()
```

```python
data = data.drop(columns=['id'])
data = data.drop(columns=['Unnamed: 32'])
```

```python
data.duplicated().sum()
```

# 03. Data Distribution (Target Variable)

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x='diagnosis', data=data)
plt.title('Distribution of Breast Cancer Classes')
plt.show()
```

# 04. Data Correlation

```python
data['diagnosis'] = data['diagnosis'].map({'M': 1, 'B': 0})
corr_matrix = data.corr()
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# 05. Prepare Features & Target for Modeling

### 01. Features & Target

```python
X = data.drop('diagnosis', axis=1)
y = data['diagnosis']
```

### 02. Standardize Features (Scaling)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### 03. Train / Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# 06. Train Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

# 07. Predictions & Evaluation

### 01. Predict on Test Set

```python
y_pred = model.predict(X_test)

y_prob = model.predict_proba(X_test)[:, 1]
```

### 02. Accuracy

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

### 03. Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(5,3))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

### 04. Classification Report

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

### 05. ROC Curve & AUC

```python
from sklearn.metrics import roc_auc_score, roc_curve

roc_auc = roc_auc_score(y_test, y_prob)
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.plot(fpr, tpr, label=f'ROC Curve (AUC={roc_auc:.2f})')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

# 08. Save Trained Model

```python
import joblib

joblib.dump(model, 'logistic_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
```
